```python
# === 1. Tautkan Google Drive ===
from google.colab import drive
drive.mount('/content/drive')

DEM_path = "/content/drive/MyDrive/KRK Imagis/DEM_Kajian.tif"
GarisPantai_path = "/content/drive/MyDrive/KRK Imagis/GarisPantai.geojson"
Sungai_path = "/content/drive/MyDrive/KRK Imagis/Sungai.geojson"

# === 2. Install dan Import Library ===
!pip install rasterio geopandas folium shapely fiona pyproj

import rasterio
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from rasterio import features
from scipy.ndimage import distance_transform_edt
import folium
import branca.colormap as cm
import matplotlib
import matplotlib.cm as mcm

# === 3. Load data DEM ===
with rasterio.open(DEM_path) as src:
    dem = src.read(1, masked=True).astype(float)
    profile = src.profile
    transform = src.transform
    crs = src.crs
    bounds = src.bounds

plt.imshow(dem, cmap="terrain")
plt.colorbar(label="Elevasi (m)")
plt.title("DEM")
plt.show()

# === 4. Load data garis pantai & sungai ===
pantai_gdf = gpd.read_file(GarisPantai_path)
sungai_gdf = gpd.read_file(Sungai_path)

# Pastikan CRS sesuai DEM
print("CRS DEM:", crs)
print("CRS Pantai:", pantai_gdf.crs)
print("CRS Sungai:", sungai_gdf.crs)

if pantai_gdf.crs != crs:
    pantai_gdf = pantai_gdf.to_crs(crs)

if sungai_gdf.crs != crs:
    sungai_gdf = sungai_gdf.to_crs(crs)

# === 5. Hitung Slope (pakai numpy gradient) ===
dx = transform[0]      # resolusi piksel x
dy = -transform[4]     # resolusi piksel y

grad_y, grad_x = np.gradient(dem, dy, dx)
slope_percent = np.sqrt(grad_x**2 + grad_y**2) * 100

plt.imshow(slope_percent, cmap="magma")
plt.colorbar(label="Slope (%)")
plt.title("Kemiringan Lereng")
plt.show()

# === 6. Rasterisasi Garis Pantai ===
coastline_mask = features.rasterize(
    ((geom, 1) for geom in pantai_gdf.geometry),
    out_shape=dem.shape,
    transform=transform,
    fill=0,
    dtype=np.uint8
)

plt.imshow(coastline_mask, cmap="gray")
plt.title("Coastline Raster Mask")
plt.show()

# Hitung jarak ke pantai
coast_dist_pixels = distance_transform_edt(1 - coastline_mask)
pixel_resolution_meters = dx  # asumsi resolusi sama
coast_dist = coast_dist_pixels * pixel_resolution_meters

plt.imshow(coast_dist, cmap="viridis")
plt.colorbar(label="Jarak ke Pantai (m)")
plt.title("Jarak ke Pantai")
plt.show()

# Klasifikasi jarak ke pantai
coast_class = np.zeros_like(coast_dist, dtype=np.uint8)
coast_class[(coast_dist < 500)] = 1
coast_class[(coast_dist >= 500) & (coast_dist < 1000)] = 2
coast_class[(coast_dist >= 1000) & (coast_dist < 2500)] = 3
coast_class[(coast_dist >= 2500) & (coast_dist < 3000)] = 4
coast_class[(coast_dist >= 3000)] = 5

plt.imshow(coast_class, cmap="Blues")
plt.colorbar(label="Kelas Jarak ke Pantai")
plt.title("Klasifikasi Jarak ke Pantai")
plt.show()
print("Max jarak pantai (m):", np.nanmax(coast_dist))

# === 7. Rasterisasi Sungai ===
river_mask = features.rasterize(
    ((geom, 1) for geom in sungai_gdf.geometry),
    out_shape=dem.shape,
    transform=transform,
    fill=0,
    dtype=np.uint8
)

plt.imshow(river_mask, cmap="gray")
plt.title("River Raster Mask")
plt.show()

river_dist_pixels = distance_transform_edt(1 - river_mask)
river_dist = river_dist_pixels * pixel_resolution_meters

plt.imshow(river_dist, cmap="viridis")
plt.colorbar(label="Jarak ke Sungai (m)")
plt.title("Jarak ke Sungai")
plt.show()

# Klasifikasi jarak ke sungai
river_class = np.zeros_like(river_dist, dtype=np.uint8)
river_class[(river_dist < 250)] = 1
river_class[(river_dist >= 250) & (river_dist < 500)] = 2
river_class[(river_dist >= 500) & (river_dist < 750)] = 3
river_class[(river_dist >= 750) & (river_dist < 1000)] = 4
river_class[(river_dist >= 1000)] = 5

plt.imshow(river_class, cmap="Greens")
plt.colorbar(label="Kelas Jarak ke Sungai")
plt.title("Klasifikasi Jarak ke Sungai")
plt.show()

# === 8. Normalisasi & Skoring Kerawanan ===
def normalize(arr):
    arr = np.where(np.isfinite(arr), arr, np.nan)
    vmin, vmax = np.nanpercentile(arr, (2, 98))
    return np.clip((arr - vmin) / (vmax - vmin), 0, 1)

elev_n = normalize(dem)
slope_n = normalize(slope_percent)
coast_n = normalize(coast_dist)
river_n = normalize(river_dist)

w_elev, w_slope, w_coast, w_river = 0.35, 0.25, 0.30, 0.10

vuln = ((1 - elev_n) * w_elev +
        (1 - slope_n) * w_slope +
        (1 - coast_n) * w_coast +
        (1 - river_n) * w_river)

plt.imshow(vuln, cmap="RdYlGn_r")
plt.colorbar(label="Skor Kerawanan (0–1)")
plt.title("Peta Skor Kerawanan Tsunami")
plt.show()

# Klasifikasi menjadi 5 kelas
classes = np.digitize(vuln, bins=[0.2, 0.4, 0.6, 0.8], right=True) + 1

plt.imshow(classes, cmap="RdYlGn_r")
plt.colorbar(label="Kelas Kerawanan (1–5)")
plt.title("Kelas Kerawanan Tsunami")
plt.show()

# === 9. Visualisasi WebGIS dengan Folium ===
lat = (bounds.top + bounds.bottom) / 2
lon = (bounds.left + bounds.right) / 2

m = folium.Map(location=[lat, lon], zoom_start=10, tiles="openstreetmap")

colors = ["#2c7bb6", "#7fcdbb", "#ffffbf", "#fdae61", "#d7191c"]
colormap = cm.StepColormap(colors, vmin=1, vmax=5, caption="Kerawanan Tsunami")

cmap = mcm.ScalarMappable(cmap=matplotlib.colors.ListedColormap(colors))
rgba_img = cmap.to_rgba(classes, bytes=True)
image = rgba_img

img_bounds = [[bounds.bottom, bounds.left], [bounds.top, bounds.right]]

folium.raster_layers.ImageOverlay(
    image=image,
    bounds=img_bounds,
    opacity=0.6,
    name="Kerawanan Tsunami"
).add_to(m)

colormap.add_to(m)
folium.LayerControl().add_to(m)

output_html = "/content/drive/MyDrive/KRK Imagis/Kerawanan_Tsunami_map.html"
m.save(output_html)
print("Saved:", output_html)

m
```
